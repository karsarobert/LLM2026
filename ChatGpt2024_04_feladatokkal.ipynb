**Érzelem-elemzés – Mesterséges intelligencia technikák az érzelmek felismerésére a szövegben**

Az érzelem-elemzés (sentiment analysis) az egyik legfontosabb, mesterséges intelligencia technikákat alkalmazó megoldás a  prompt engineering (a természetes nyelvre épülő, mesterséges intelligenciával vezérelt feladatok automatizálásának) területén. Ez az első az általunk tárgyalt AI-alkalmazások sorában. Az érzelem-elemzés természetes nyelvi feldolgozást (Natural Language Processing, NLP) használ az érzelmi jelek azonosítására és kinyerésére a szöveges adatokból. A cél az, hogy felmérjük egy írásmű általános érzelmét, például meghatározzuk, hogy egy termék értékelése pozitív vagy negatív-e.

**Az érzelem-elemzés leggyakoribb felhasználási területei a következők:**

* **Márkamonitorozás:** A márkák nyomon követhetik az online zajló beszélgetéseket, hogy figyelemmel kísérjék hírnevüket. Az érzelem-elemzés lehetővé teszi számukra, hogy felismerjék a felmerülő PR-problémákat, és mérjék a marketingkampányok hatását.
* **Ügyfélszolgálat:** Az ügyfélélményért felelős csapatok nagyszabásúan összesíthetik a visszajelzéseket felmérések, vélemények és közösségi média említések elemzésével. Ez adatvezérelt betekintést nyújt abba, hogy az emberek hol elégedettek vagy csalódottak a termékekkel és szolgáltatásokkal.
* **Piackutatás:** Az ügyfelek hangulatának megértésével a vállalkozások felfedezhetik szolgáltatásaik vagy termékeik hiányosságait, és ennek megfelelően pótolhatják azokat.
* **Politikai elemzés:** A politikai csoportok szemmel tarthatják a választók hozzáállását a jelöltekkel és a problémákkal kapcsolatos viták nyomon követésével. A technológia segít az üzenetek tesztelésében és azoknak a területeknek a meghatározásában, ahol további felvilágosításra van szükség.
* **Közösségi média megfigyelése:** A vállalatok az érzelem-elemzést használják fel a közösségi média bejegyzéseinek hangulatának azonosítására, hogy megértsék a szélesebb fogyasztói magatartást és trendeket.
* **Befektetési elemzés:** A befektetők érzelem-elemzést használhatnak a piaci mozgások előrejelzéséhez. A releváns hírek, fórumok és blogok feldolgozásával az eszközök meg tudják határozni, hogy a befektetők optimisták vagy pesszimisták-e bizonyos részvényekkel vagy szektorokkal kapcsolatban.
* **Közegészségügy:** Az egészségügyi hatóságok érzelem-elemzést alkalmazhatnak a közvélemény aggályainak, a félretájékoztatásnak és az érzelmi állapotoknak a megértésére olyan válságok idején, mint például a világjárványok. #Adattisztítás - avagy hogyan azonosít és orvosol az AI problémákat az adathalmazokban

Az adattisztítás egy olyan folyamat, amely során beazonosítják és korrigálják (vagy eltávolítják) a hibákat, eltéréseket és pontatlanságokat egy adathalmazból. Ez magában foglalhatja a duplikált rekordok észlelését, hiányos vagy hiányzó adatok kezelését, értékek ellenőrzését és javítását, az adatok formázását, szabványosítását vagy normalizálását, valamint a kiugró értékek kezelését.

Az adattisztítás fő célja az adatok minőségének és megbízhatóságának javítása, annak biztosítása, hogy azok pontosak, konzisztensek és a felhasználási célra alkalmas formátumúak legyenek. Ez azért döntő fontosságú, mert a hiányos, rendetlen (angolul: *dirty*) adatok zavarhatják az elemzéseket, pontatlan felismerésekhez és következtetésekhez vezethetnek, és negatív hatással lehetnek a döntéshozatali folyamatokra.

Fontos megjegyezni, hogy az adattisztításnak az adatkezelés rendszeres részének kell lennie – nem egyszeri feladat, mivel új hibák keletkezhetnek, amikor új adatok kerülnek hozzáadásra vagy a meglévő adatok módosulnak.

Általánosságban elmondható, hogy az adattisztítás összetett és költséges folyamat volt, amelyet generatív AI-val hatékonyan lehetséges elvégezni.

```python
import google.generativeai as genai
from google.colab import userdata
import json

# API kulcs betöltése Colab Secrets-ből (biztonságos)
try:
    api_key = userdata.get('GEMINI_API_KEY')
    print("API kulcs sikeresen betöltve a Colab Secrets-ből.")
except Exception:
    import getpass
    api_key = getpass.getpass('Add meg a Gemini API kulcsot: ')

genai.configure(api_key=api_key)

# Modell inicializálása (gemini-1.5-flash: ingyenes tierben elérhető, gyors)
model = genai.GenerativeModel('gemini-2.5-flash')
print(f"Modell kész: {model.model_name}")
```

### Érzelem-elemzés Gemini API-val

Az előző fejezetben kézzel írtuk be a promptot a ChatGPT-be. Most ugyanezt **Python kódból** hívjuk meg.

```python
def sentiment_elemzes(szoveg: str) -> str:
    """Egyetlen mondat érzelem-elemzése. Visszaadja: pozitív / semleges / negatív"""
    prompt = f"""Azonosítsd és kategorizáld az alábbi szövegben kifejezett érzelmet.
Csak egy szóban válaszolj: pozitív, semleges, vagy negatív.

Szöveg: {szoveg}"""
    response = model.generate_content(prompt)
    return response.text.strip()

# Tesztelés
teszt_mondatok = [
    "Ez a termék csodálatos! Tényleg javított az életemen.",
    "A projektjavaslatom beadási határideje jövő pénteken 17 órakor lejár.",
    "Nagyon csalódott vagyok a szörnyű ügyfélszolgálat miatt.",
    "Még ma be kell ugranom az élelmiszerboltba tejért és tojásért.",
]

print(f"{'Szöveg (rövidítve)':<55} | {'Hangulat'}")
print("-" * 70)
for mondat in teszt_mondatok:
    eredmeny = sentiment_elemzes(mondat)
    print(f"{mondat[:52]+'...':<55} | {eredmeny}")
```

### Tömeges érzelem-elemzés – JSON kimenet

A Gemini `response_mime_type="application/json"` beállítással garantáltan valid JSON-t ad vissza, ami közvetlenül feldolgozható.

```python
mondatok = [
    "Nagyon csalódott vagyok a szörnyű ügyfélszolgálat miatt, amit kaptam.",
    "Az ételek ebben az étteremben abszolút finomak. Mindenképpen vissza fogok térni.",
    "Nemrég vettem ezt a laptopot, és ez a legjobb technikai vásárlás évek óta.",
    "Úgy néz ki, hogy ma délután esni fog. Vinnem kéne egy esernyőt.",
    "A projektjavaslatom beadási határideje jövő pénteken 17 órakor lejár.",
    "Most néztem meg az új Marvel filmet – teljesen felülmúlta az elvárásaimat.",
    "Ha manapság a híreket olvasom, csak szorongok és lehangolt vagyok.",
    "Ez a legrosszabb ízű étel, amit valaha ettem egy étteremben.",
]

szovegek_str = "\n".join([f'{i+1}. "{m}"' for i, m in enumerate(mondatok)])

prompt = f"""Azonosítsd és kategorizáld az alábbi mondatokban kifejezett érzelmeket.
Csak 'pozitív', 'semleges' vagy 'negatív' értékeket használj.
Válaszolj JSON formátumban: egy "eredmenyek" kulcsú lista, ahol minden elem tartalmaz
"sorszam", "szoveg" és "hangulat" mezőket.

Mondatok:
{szovegek_str}"""

response = model.generate_content(
    prompt,
    generation_config=genai.GenerationConfig(
        response_mime_type="application/json"
    )
)

eredmenyek = json.loads(response.text)["eredmenyek"]

# Táblázatos megjelenítés
print(f"{'#':<4} | {'Hangulat':<12} | Szöveg")
print("-" * 80)
for e in eredmenyek:
    print(f"{e['sorszam']:<4} | {e['hangulat']:<12} | {e['szoveg'][:55]}...")
```

### Teljes pipeline: Tesseract OCR + Gemini

A folyamat lépései kódban:
1. Képelőfeldolgozás (OpenCV) (ezt kihagyjuk)
2. Szöveg kinyerése (Tesseract)
3. Strukturálás (Gemini) → JSON

```python
def szamla_feldolgozas_ocr_gemini(kep_ut: str) -> dict:
    """
    Teljes számla-feldolgozó pipeline:
      Kép → OpenCV előfeldolgozás → Tesseract OCR → Gemini → JSON
    """
    # 1. lépés: képelőfeldolgozás + OCR
    ocr_szoveg = pytesseract.image_to_string(kep_ut, lang="hun")
    print(f"[1/2] OCR kész – {len(ocr_szoveg)} karakter kinyerve")

    # 2. lépés: Gemini strukturálja az adatokat
    prompt = f"""Az alábbi szöveg egy OCR-rel beolvasott számláról/nyugtáról származik,
ezért tartalmazhat helyesírási hibákat és zajt.

Keresd meg és nyerd ki a következő adatokat JSON formátumban:
- szallito: nev, cim, adoszam
- vevo: nev, cim, adoszam  (ha nincs, értéke null)
- szamla_sorszam            (ha nincs, értéke null)
- szamla_kelte              (ha nincs, értéke null)
- tetelek: lista – minden elemben: megnevezes, mennyiseg, egysegar, osszeg
- fizetendo_osszeg          (csak szám, Ft jel nélkül)
- fizetesi_mod              (ha nincs, értéke null)

Ha valamit nem találsz, az értéke legyen null. Csak valid JSON-t adj vissza!

OCR szöveg:
{ocr_szoveg}"""

    response = model.generate_content(
        prompt,
        generation_config=genai.GenerationConfig(
            response_mime_type="application/json"
        )
    )
    print("[2/2] Gemini strukturálás kész")
    return json.loads(response.text)

# Teszt
eredmeny = szamla_feldolgozas_ocr_gemini("Aldi.jpg")
print("\n--- Kinyert adatok ---")
print(json.dumps(eredmeny, indent=2, ensure_ascii=False))
```

### Vision LLM megközelítés – Tesseract nélkül

A **Gemini 2.5 Flash** képes közvetlenül képből olvasni – nem kell Tesseract.
Ez pontosabb, gyorsabb, és kevesebb kód.

**Mikor melyiket érdemes használni?**

| Megközelítés | Előny | Hátrány |
|---|---|---|
| Tesseract + Gemini | Ingyenes OCR, offline is futtatható | Zajos OCR ront a minőségen |
| Gemini Vision | Pontosabb, egyszerűbb kód | API hívás száma korlátozott |

```python
def szamla_feldolgozas_vision(kep_ut: str) -> dict:
    """
    Vision LLM megközelítés: a kép közvetlenül megy a Gemini-be, Tesseract nélkül.
    """
    kep = Image.open(kep_ut)

    prompt = """Olvasd be ezt a számlát vagy nyugtát és nyerd ki az adatokat JSON formátumban:
- szallito: nev, cim, adoszam
- vevo: nev, cim, adoszam  (ha nincs, értéke null)
- szamla_sorszam            (ha nincs, értéke null)
- szamla_kelte              (ha nincs, értéke null)
- tetelek: lista – minden elemben: megnevezes, mennyiseg, egysegar, osszeg
- fizetendo_osszeg          (csak szám, Ft jel nélkül)
- fizetesi_mod              (ha nincs, értéke null)

Ha valamit nem találsz, az értéke legyen null. Csak valid JSON-t adj vissza!"""

    response = model.generate_content(
        [prompt, kep],
        generation_config=genai.GenerationConfig(
            response_mime_type="application/json"
        )
    )
    return json.loads(response.text)

# Összehasonlítás a két megközelítés között
print("=" * 50)
print("Tesseract + Gemini pipeline:")
print("=" * 50)
eredmeny_ocr = szamla_feldolgozas_ocr_gemini("Aldi.jpg")
print(json.dumps(eredmeny_ocr, indent=2, ensure_ascii=False))

print("\n" + "=" * 50)
print("Gemini Vision (közvetlen):")
print("=" * 50)
eredmeny_vision = szamla_feldolgozas_vision("Aldi.jpg")
print(json.dumps(eredmeny_vision, indent=2, ensure_ascii=False))
```

### Kötegelt feldolgozás – több számla egyszerre

A sofőrök naponta több számlát küldenek be. Az alábbi kód **automatikusan feldolgoz minden képet** és összesíti az eredményeket egy táblázatba.

```python
import pandas as pd

def kotegelt_feldolgozas(kep_lista: list, modszer: str = "vision") -> pd.DataFrame:
    """
    Több számla/nyugta feldolgozása és összesítése DataFrame-be.
    modszer: 'vision' (Gemini Vision) vagy 'ocr' (Tesseract + Gemini)
    """
    feldolgozo = szamla_feldolgozas_vision if modszer == "vision" else szamla_feldolgozas_ocr_gemini
    eredmenyek = []

    for kep_ut in kep_lista:
        print(f"Feldolgozás: {kep_ut} ...", end=" ")
        try:
            adat = feldolgozo(kep_ut)
            adat["forras_fajl"] = kep_ut
            eredmenyek.append(adat)
            print("OK")
        except Exception as e:
            print(f"HIBA: {e}")
            eredmenyek.append({"forras_fajl": kep_ut, "hiba": str(e)})

    return pd.DataFrame(eredmenyek)

# Összes letöltött számla feldolgozása
kepek = ["Aldi.jpg", "szamla1.jpg", "szamla.jpg"]
df = kotegelt_feldolgozas(kepek, modszer="vision")

# Összesített nézet
print("\n--- Összesítő táblázat ---")
oszlopok = [c for c in ["forras_fajl", "szamla_sorszam", "szamla_kelte", "fizetendo_osszeg"] if c in df.columns]
print(df[oszlopok].to_markdown(index=False))

# Mentés CSV-be
df.to_csv("szamlak_osszesito.csv", index=False, encoding="utf-8-sig")
print("\nEredmény elmentve: szamlak_osszesito.csv")
```
