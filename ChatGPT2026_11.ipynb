## 1. Bevezetés és Célkitűzés

### Mi a LangChain?

A **LangChain** egy keretrendszer, amelyet arra terveztek, hogy megkönnyítse a **nagy nyelvi modelleket (LLM-eket)** használó alkalmazások fejlesztését. A neve onnan ered, hogy különböző komponenseket "láncokká" (chains) lehet összekapcsolni.

**Miért van rá szükség?**

Amikor egy LLM-mel dolgozunk, gyakran nem elég csak egy kérdést feltenni és megkapni a választ. Sokszor szükség van:
- Külső dokumentumok betöltésére és feldolgozására
- Releváns információk keresésére nagy mennyiségű szövegben
- Többlépéses gondolkodásra (pl. először összegyűjt információkat, majd válaszol)
- Memóriakezelésre (korábbi beszélgetések megjegyzésére)

A LangChain mindehhez **építőkockákat** biztosít, amelyeket könnyen össze tudunk kapcsolni.

### A gyakorlat célja

Ez a gyakorlat bemutatja a LangChain alapvető építőköveit:
1. **Modellek**: Hogyan kommunikáljunk LLM-ekkel
2. **Promptok**: Hogyan készítsünk dinamikus utasításokat
3. **Dokumentum betöltők**: Hogyan dolgozzunk fel különböző fájlokat
4. **Szövegdarabolás**: Hogyan vágjuk fel a hosszú szövegeket
5. **Beágyazások**: Hogyan alakítsuk a szöveget számokká
6. **Vektortárolók**: Hogyan keressünk szemantikus hasonlóság alapján
7. **Láncok**: Hogyan kapcsoljuk össze a komponenseket

### 3.1. Modellek (LLMs)

**Leírás:** A nyelvi modellek a LangChain szíve. Ők azok, akik a szöveget generálják, kérdésekre válaszolnak, fordítanak stb.

**Groq előnyei:**
- Rendkívül gyors válaszidő (speciális hardver)
- Ingyenes API használat korlátozott mértékben
- OpenAI-kompatibilis API (könnyű használat)

**Modell beállítás:**

```python
from openai import OpenAI

# OpenAI kliens létrehozása Groq-ra irányítva
# A 'base_url' paraméterrel adjuk meg, hogy a Groq API-t használjuk
client = OpenAI(
    base_url='https://api.groq.com/openai/v1',  # Groq API végpont
    api_key=api_key
)

# A használandó modell
MODELL = 'openai/gpt-oss-120b'

print(f'✓ OpenAI kliens létrehozva Groq-kal')
print(f'✓ Modell: {MODELL}')
```

**LangChain integráció:**



```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    base_url="https://api.groq.com/openai/v1",
    api_key=api_key,
    model="openai/gpt-oss-120b",
    temperature=0.7
)

response = llm.invoke("Mondj egy viccet a programozásról!")

print("\nModell válasza:")
print(response.content)
```

**Magyarázat:**

- A `ChatOpenAI` egy chat-alapú LangChain modellobjektum.
- A `base_url` paraméterrel nem az OpenAI saját végpontját, hanem a **Groq OpenAI-kompatibilis API-ját** használjuk.
- A `model` paraméter adja meg, melyik modellt szeretnénk futtatni.
- A `temperature` a válaszok változatosságát szabályozza:
  - `0.0`: determinisztikusabb, kiszámíthatóbb válaszok
  - `0.7`: természetesebb, kreatívabb válaszok
  - `1.0` körül: nagyon változatos, kevésbé kontrollált kimenet

A `llm.invoke(...)` hívás eredménye egy **AIMessage** objektum, ezért a tényleges szöveg általában a `.content` mezőben érhető el.

### 3.2. Prompt Sablonok (Prompt Templates)

**Leírás:** A prompt sablonok segítenek **dinamikus**, újrafelhasználható utasításokat létrehozni. Gondolj rá úgy, mint egy üzenetmintára, amibe változókat helyettesíthetsz be.

**Miért használjuk?**
- Nem kell minden alkalommal újraírni a promptot
- Könnyen változtatható a bemeneti adat
- Tisztább, karbantarthatóbb kód

**Példa:**

```python
from langchain_core.prompts import PromptTemplate

# Sablon létrehozása
template = """Fordítsd le az alábbi mondatot {nyelvre}.
Csak a fordítást add meg.

Mondat: {mondat}
Fordítás:"""

# PromptTemplate objektum
prompt = PromptTemplate(
    template=template,
    input_variables=["mondat", "nyelvre"]  # Ezeket a változókat várjuk
)

# Prompt formázása konkrét értékekkel
formatted_prompt = prompt.format(
    mondat="the cat is on the table",
    nyelvre="magyarra"
)

print('Formázott prompt:')
print(formatted_prompt)
print('\n' + '='*50)

# Használat az LLM-mel
valasz = llm.invoke(formatted_prompt)
print('\nLLM válasza:')
print(valasz.content)
```

**Hogyan működik?**

1. Megadjuk a sablont `{változó}` jelölésekkel.
2. A `PromptTemplate` ezeket a változókat kezeli.
3. A `.format()` metódussal konkrét értékekre cseréljük őket.
4. Az így kapott szöveg már közvetlenül elküldhető a modellnek.

A prompt sablonok előnye, hogy ugyanazt az utasításszerkezetet több különböző bemenettel is újra tudjuk használni.

### 3.4. Dokumentum Átalakítók - Szövegdarabolás (Text Splitters)

**Probléma:** Az LLM-eknek van egy maximum kontextusméretük (pl. 8000, 32000 token). Egy hosszú dokumentum nem fér bele egyszerre.

**Megoldás:** Daraboljuk fel kisebb "chunk"-okra (darabokra).

**Fontos kérdések:**
- Milyen nagy legyen egy darab? → `chunk_size`
- Legyen-e átfedés a darabok között? → `chunk_overlap`
  - Az átfedés megőrzi a kontextust a darabhatárokon

#### RecursiveCharacterTextSplitter

Ez a legokosabb szövegdaraboló. Megpróbálja **logikai egységek mentén** (bekezdések, mondatok) vágni a szöveget.

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Szövegdaraboló létrehozása
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,        # Max. 200 karakter egy darabban
    chunk_overlap=50,      # 50 karakter átfedés
    length_function=len,   # Hogyan mérjük a hosszt (karakterszám)
)

# Dokumentum darabolása
chunks = text_splitter.split_documents(text_documents)

print(f'✓ Eredeti dokumentumok száma: {len(text_documents)}')
print(f'✓ Darabolás után: {len(chunks)} darab')
print(f'\n{"="*60}')

# Első 3 darab megtekintése
for i in range(min(5, len(chunks))):
    print(f'\n{i+1}. darab ({len(chunks[i].page_content)} karakter):')
    print(chunks[i].page_content)
    print('-' * 60)
```

**Mit látunk?**

- A hosszú dokumentumot a splitter kisebb darabokra bontja.
- Az egymást követő darabok között **átfedés** van, így fontos szövegrészek nem vesznek el a vágási pontokon.
- A `RecursiveCharacterTextSplitter` megpróbál értelmes helyeken vágni, például sortörések vagy bekezdéshatárok mentén.
- Ez különösen fontos RAG rendszerekben, mert a visszakeresés jellemzően ezeken a kisebb dokumentumdarabokon történik.

A darabolás célja, hogy a dokumentumok jól kezelhetők maradjanak embedding és retrieval során is.

### 3.5. Szövegbeágyazó Modellek (Text Embedding Models)

**Mi az a beágyazás (embedding)?**

A szöveg → szám (vektor) átalakítás. Pl.:
- "kutya" → [0.2, 0.8, -0.3, 0.1, ...] (784 szám)
- "macska" → [0.25, 0.75, -0.28, 0.15, ...] (784 szám)

**Miért hasznos?**

- A hasonló jelentésű szavak/mondatok **hasonló vektorokat** kapnak
- Számítógép számára könnyű kiszámítani, hogy két vektor mennyire "közeli"
- Ez lehetővé teszi a **szemantikus keresést** (jelentés alapján, nem csak szó egyezés)

**Példa:**

Használjuk a `microsoft/harrier-oss-v1-0.6b` modellt, amely:
- Kis méretű (600M paraméter)
- Gyors
- Jó minőségű beágyazásokat ad
- 1024 dimenziós vektorokat generál

```python
from sentence_transformers import SentenceTransformer

# Beágyazó modell betöltése (első futáskor letölti)
embedding_model = SentenceTransformer('microsoft/harrier-oss-v1-0.6b')

print('✓ Beágyazó modell betöltve')
print(f'✓ Modell neve: microsoft/harrier-oss-v1-0.6b')
print(f'✓ Vektor dimenzió: {embedding_model.get_sentence_embedding_dimension()}')
```

**LangChain integráció:**

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

# LangChain-kompatibilis beágyazó
embeddings = HuggingFaceEmbeddings(
    model_name='microsoft/harrier-oss-v1-0.6b'
)

# Példa szövegek beágyazása
documents_to_embed = [
    "Jó reggelt!",
    "Szia!",
    "Balesetet szeretnék bejelenteni",
    "Megkérdezhetem a nevét?",
    "Persze, Nagy János vagyok."
]

# Dokumentumok beágyazása
doc_embeddings = embeddings.embed_documents(documents_to_embed)

print('\nDokumentum beágyazások:')
print(f'✓ Beágyazott dokumentumok száma: {len(doc_embeddings)}')
print(f'✓ Egy vektor dimenziója: {len(doc_embeddings[0])}')

# Lekérdezés beágyazása (egy kérdés)
query = "Mi volt a bejelentő neve?"
query_embedding = embeddings.embed_query(query)

print(f'\nLekérdezés beágyazása:')
print(f'✓ Vektor dimenziója: {len(query_embedding)}')
print(f'✓ Első 5 elem: {query_embedding[:5]}')
```

**Különbség:**

- `embed_documents()`: Több szöveg beágyazása (listát ad vissza)
- `embed_query()`: Egy lekérdezés beágyazása (egyetlen vektort ad vissza)

**Hasonlóság számítás kézzel:**

Nézzük meg, melyik dokumentum hasonlít leginkább a kérdésre!

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Koszinusz hasonlóság számítása
similarities = cosine_similarity(
    [query_embedding],  # A lekérdezés vektora
    doc_embeddings      # Az összes dokumentum vektora
)[0]

print('\nHasonlóságok (0=eltérő, 1=azonos):')
print('='*60)
for i, (doc, sim) in enumerate(zip(documents_to_embed, similarities)):
    print(f'{i+1}. [{sim:.4f}] "{doc}"')

# Legrelevánsabb dokumentum
best_idx = np.argmax(similarities)
print(f'\n✓ Legrelevánsabb: "{documents_to_embed[best_idx]}"')
```

**Magyarázat:**

A koszinusz hasonlóság azt méri, hogy két vektor mennyire "mutat ugyanabba az irányba".
- 1.0 = teljesen azonos irány (nagyon hasonló)
- 0.0 = merőleges (semmi közük egymáshoz)
- -1.0 = ellentétes irány (ellentétes jelentés)

### 3.6. Vektortárolók (Vector Stores)

**Probléma:** Ha sok ezer dokumentumunk van, akkor minden kereséskor újraszámolni az összes hasonlóságot lassú lenne.

**Megoldás:** Vektortárolók (vector stores/vector databases)

- Tárolják a vektorokat
- Gyors hasonlósági keresést biztosítanak
- Automatikusan megtalálják a legközelebbi vektorokat

**FAISS (Facebook AI Similarity Search):**
- Meta (Facebook) által fejlesztett
- Rendkívül gyors
- CPU és GPU támogatás
- Akár milliárd vektort is kezel

#### Példa: Dialog.txt feldolgozása

Töltsük le a példa fájlt:

```python
!gdown 1tICd5l-gv1PhZgvCkLD4ydqq-jIaHUv-
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.vectorstores import FAISS

# 1. Dokumentum betöltése
raw_documents = TextLoader('dialog.txt', encoding='utf-8').load()

# 2. Darabolás (soronként)
text_splitter = CharacterTextSplitter(
    chunk_size=50,
    chunk_overlap=0,
    separator="\n"  # Soronként vágunk
)
documents = text_splitter.split_documents(raw_documents)

print(f'✓ Dokumentumok száma: {len(documents)}')
print('\nDokumentumok:')
for i, doc in enumerate(documents):
    print(f'{i+1}. "{doc.page_content}"')
```

**FAISS vektortároló létrehozása:**

```python
# 3. Vektortároló létrehozása
# Ez automatikusan:
# - Beágyazza az összes dokumentumot
# - Létrehoz egy FAISS indexet
# - Tárolja a dokumentumokat és a vektorokat

db = FAISS.from_documents(documents, embeddings)

print('✓ FAISS vektortároló létrehozva')
print(f'✓ Tárolt dokumentumok száma: {len(documents)}')
```

**Hasonlósági keresés:**

```python
# 4. Keresés
query = "What is the reason for calling?"

# Alapértelmezetten 4 legközelebbi dokumentumot ad vissza
results = db.similarity_search(query, k=4)

print(f'Kérdés: "{query}"')
print('\nLeginkább releváns dokumentumok:')
print('='*60)
for i, doc in enumerate(results):
    print(f'{i+1}. "{doc.page_content}"')
```

**Keresés hasonlósági pontszámmal:**

Nézzük meg, mennyire releváns mindegyik találat:

```python
# Keresés pontszámokkal
results_with_scores = db.similarity_search_with_score(query, k=4)

print(f'Kérdés: "{query}"')
print('\nTalálatok pontszámokkal:')
print('='*60)
for i, (doc, score) in enumerate(results_with_scores):
    print(f'{i+1}. [távolság: {score:.4f}] "{doc.page_content}"')

print('\n💡 Alacsonyabb távolság = relevánsabb találat')
```

**Magyarázat:**

- A FAISS egy vektortároló, amely gyors hasonlósági keresést tesz lehetővé.
- A dokumentumokat először embedding vektorokká alakítjuk.
- Kereséskor a lekérdezést is beágyazzuk, majd a rendszer megkeresi a hozzá legközelebb eső dokumentumvektorokat.
- A `similarity_search()` a legrelevánsabb találatokat adja vissza.
- A `similarity_search_with_score()` esetén a találatokhoz pontszám is tartozik.

Így nem kell minden keresésnél kézzel végigszámolni az összes dokumentum hasonlóságát.

### 3.7. Lekérdezők (Retrievers)

**Leírás:** A retrieverek egy magasabb szintű absztrakciót adnak a dokumentum-visszakereséshez.  
Míg a vektortároló közvetlen keresési műveleteket biztosít, a retriever egységes interfészt ad arra, hogyan kérjünk vissza releváns dokumentumokat.

A modern LangChain-ben a retrieverek a **Runnable** szemlélethez illeszkednek, ezért jellemzően `invoke()` hívással használjuk őket.

**Mikor hasznos?**
- ha később láncba akarjuk kötni a visszakeresést,
- ha a keresési logikát könnyen cserélhetővé akarjuk tenni,
- ha RAG rendszerhez szeretnénk előkészíteni a dokumentumokat.

```python
# Retriever létrehozása a vektortárolóból
retriever = db.as_retriever(
    search_type="similarity",  # Hasonlósági keresés
    search_kwargs={"k": 3}      # 3 találatot kérünk
)

print('✓ Retriever létrehozva')
print(f'  Keresési típus: similarity')
print(f'  Visszaadott találatok száma: 3')
```

**Használat:**

```python
query = "What was the reason of the call?"

# Releváns dokumentumok lekérése
relevant_docs = retriever.invoke(query)

print(f'Kérdés: "{query}"')
print('\nRetriever által visszaadott dokumentumok:')
print('='*60)
for i, doc in enumerate(relevant_docs):
    print(f'{i+1}. "{doc.page_content}"')
```

**Különbség a közvetlen FAISS kereséshez képest:**

- A FAISS közvetlenül a vektortároló keresési műveleteit adja.
- A retriever egy egységesebb, magasabb szintű interfész.
- A retriever könnyebben beépíthető nagyobb LangChain folyamatokba.
- A modern használatban a retrievert jellemzően `invoke()` metódussal hívjuk meg.

Ezért a retriever különösen hasznos RAG és összetettebb láncok építésekor.

## 4. Láncok és Runnable-ok: Komponensek Összekapcsolása

**Mi a cél?**

Eddig minden komponenst külön-külön használtunk. A következő lépés az, hogy ezeket **összekapcsoljuk**.

A modern LangChain-ben ez gyakran úgy történik, hogy:
- promptot készítünk,
- modellt kapcsolunk hozzá,
- majd opcionálisan egy output parserrel egyszerűsítjük a kimenetet.

Ezt a szemléletet sokszor **LCEL**-nek vagy egyszerűen **runnable láncolásnak** is nevezik.

```python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI

template_translate = """Mondat: {mondat}
Fordítás {nyelvre}:"""

prompt_translate = PromptTemplate.from_template(template_translate)

llm_fordito = ChatOpenAI(
    base_url="https://api.groq.com/openai/v1",
    api_key=api_key,
    model="openai/gpt-oss-120b",
    temperature=0
)

fordito_lanc = prompt_translate | llm_fordito

print("✓ Fordító lánc létrehozva")

valasz = fordito_lanc.invoke({
    "mondat": "the cat is on the table",
    "nyelvre": "magyarra"
})

print(valasz.content)
```

### 4.1. Egyszerű runnable lánc – Prompt + modell

**Leírás:**  
Az egyik legegyszerűbb lánctípus az, amikor egy prompt sablont közvetlenül összekötünk egy modellel.

Ebben a példában:
- a bemenet egy mondat és egy célnyelv,
- a prompt sablon ebből utasítást készít,
- a modell pedig elvégzi a fordítást.

Ez a modern LangChain szemléletben a `prompt | model` mintával valósul meg.

```python
eredmeny = fordito_lanc.invoke({
    "mondat": "the cat is on the table",
    "nyelvre": "magyarra"
})

print('Eredeti mondat: "the cat is on the table"')
print(f'Fordítás magyarra: "{eredmeny.content.strip()}"')

eredmeny2 = fordito_lanc.invoke({
    "mondat": "I love programming",
    "nyelvre": "németre"
})

print(f'\nEredeti mondat: "I love programming"')
print(f'Fordítás németre: "{eredmeny2.content.strip()}"')
```

**Hogyan működik?**

1. A lánc megkapja a bemenetet szótár formában.
2. A `PromptTemplate` kitölti a sablont a kapott változókkal.
3. A `ChatOpenAI` elküldi a kész promptot a modellnek.
4. A modell válasza egy **AIMessage** objektum.
5. A tényleges válaszszöveget a `.content` mezőből olvassuk ki.

Ez a minta jól használható minden olyan feladatra, ahol egy strukturált utasítást szeretnénk újra és újra különböző adatokkal lefuttatni.

### 4.2. Többlépcsős runnable lánc – két lépés összefűzése

**Leírás:**  
Sok feladat nem egyetlen modellhívásból áll. Gyakori, hogy az egyik lépés kimenetét tovább szeretnénk adni egy következő lépésnek.

Ebben a példában:
1. először generálunk egy vicces szöveget egy adott témáról,
2. majd ezt a kimenetet magyarra fordítjuk.

A modern LangChain-ben ezt több runnable egymás utáni összekapcsolásával oldjuk meg.

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableLambda

llm_kreativ = ChatOpenAI(
    base_url="https://api.groq.com/openai/v1",
    api_key=api_key,
    model="openai/gpt-oss-120b",
    temperature=0.8
)

template_vicc = """Te egy humorista vagy. Generálj egy vicces történetet a következő témáról angolul: {tema}

Vicc:"""

prompt_vicc = PromptTemplate.from_template(template_vicc)

template_forditas = """Te egy fordító vagy. Fordítsd le magyarra a következő szöveget:

{input}

Magyar fordítás:"""

prompt_forditas = PromptTemplate.from_template(template_forditas)

vicc_lanc = prompt_vicc | llm_kreativ | StrOutputParser()
forditas_lanc = prompt_forditas | llm_kreativ | StrOutputParser()

teljes_lanc = (
    vicc_lanc
    | RunnableLambda(lambda text: {"input": text})
    | forditas_lanc
)

#teljes_lanc = (
#    prompt_vicc
#    | llm_kreativ
#    | StrOutputParser()
#    | {"input": RunnablePassthrough()} #a RunnablePassthrough() változtatás nélkül továbbadja az előző lépés kimenetét.
#    | prompt_forditas
#    | llm_kreativ
#    | StrOutputParser()
#)

print("✓ Szekvenciális lánc létrehozva")

eredmeny = teljes_lanc.invoke({"tema": "programozók és kávé"})
print(eredmeny)
```

**Futtatás:**

```python
# Futtatás
eredmeny = teljes_lanc.invoke({"tema":"macskák és kutyák"})

print('\n' + '='*60)
print('VÉGEREDMÉNY:')
print('='*60)
print(eredmeny)
```

**Mit látunk?**

- Az első runnable lánc elkészíti a vicces szöveget.
- Ezután a köztes kimenetet átalakítjuk olyan formára, amelyet a következő prompt már bemenetként tud fogadni.
- A második runnable lánc lefordítja ezt a szöveget magyarra.
- Az egész folyamatot egyetlen `invoke()` hívással futtatjuk végig.

Ez jól mutatja, hogy egy összetettebb feladatot több kisebb, jól elkülöníthető lépésre bonthatunk.

### 4.3. RetrievalQA - RAG (Retrieval-Augmented Generation)

**Mi az a RAG?**

**R**etrieval-**A**ugmented **G**eneration = Visszakereséssel kiegészített generálás

**Probléma:** Az LLM-ek nem tudnak mindent, és a tudásuk egy adott időpontban "befagyott".

**Megoldás:** RAG
1. Keress releváns dokumentumokat (retrieval)
2. Add oda az LLM-nek kontextusként
3. Az LLM a dokumentumok alapján válaszol (generation)

**Előnyök:**
- Az LLM naprakész információkkal dolgozik
- Saját dokumentumainkra tud válaszolni
- Csökkenti a "hallucinációt" (kitalált válaszokat)

#### Példa: Kérdés-válasz a Pál Utcai Fiúk alapján

Letöltjük a könyvet:

```python
!gdown 1op1icuNvSOmXzfrfFPfGGsi-BX4T6rNI
```

**RAG pipeline felépítése:**

1. betöltjük a dokumentumot,
2. kisebb darabokra bontjuk,
3. embeddingeket készítünk belőlük,
4. eltároljuk őket egy vektortárolóban,
5. létrehozunk egy retrievert,
6. a releváns darabokat egy promptba illesztjük,
7. a modell ezek alapján válaszol.

A modern LangChain-ben ezt gyakran külön retrieverrel és külön dokumentum-összefűző lánccal építjük fel.

```python
# Ha kell:
# !pip install -U langchain-openai langchain-community langchain-text-splitters langchain-classic faiss-cpu

from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_classic.chains import create_retrieval_chain
from langchain_classic.chains.combine_documents import create_stuff_documents_chain

print("RAG Pipeline építése...")
print("=" * 60)

# 1. Dokumentum betöltése
loader_puf = TextLoader("PUF.txt", encoding="utf-8")
raw_documents_puf = loader_puf.load()
print(f'✓ Betöltve: {len(raw_documents_puf[0].page_content)} karakter')

# 2. Darabolás
text_splitter_puf = CharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    separator="\n"
)
documents_puf = text_splitter_puf.split_documents(raw_documents_puf)
print(f'✓ Darabok száma: {len(documents_puf)}')

# 3. Vektortároló
db_puf = FAISS.from_documents(documents_puf, embeddings)
print(f'✓ FAISS index létrehozva {len(documents_puf)} vektorral')

# 4. Retriever
retriever_puf = db_puf.as_retriever(search_kwargs={"k": 4})
print("✓ Retriever kész")

# 5. LLM
llm_puf = ChatOpenAI(
    base_url="https://api.groq.com/openai/v1",
    api_key=api_key,
    model="openai/gpt-oss-120b",
    temperature=0
)
print("✓ LLM kész")

# 6. QA chain
prompt = ChatPromptTemplate.from_messages([
    ("system", "Válaszolj kizárólag a megadott kontextus alapján. Ha nincs benne a válasz, mondd ezt világosan.\n\nKontextus:\n{context}"),
    ("human", "{input}")
])

combine_docs_chain = create_stuff_documents_chain(llm_puf, prompt)
qa_chain = create_retrieval_chain(retriever_puf, combine_docs_chain)
print("✓ QA lánc kész")

print("\n" + "=" * 60)
print("✓ RAG Pipeline kész!")
print("=" * 60)

# Kérdés
result = qa_chain.invoke({"input": "Mi a PUF lényege?"})

print("\nVálasz:")
print(result["answer"])
```

**Mi történik ebben a megoldásban?**

- A retriever kiválasztja a kérdéshez legrelevánsabb dokumentumdarabokat.
- A dokumentumdarabokat egy külön lánc összefűzi és behelyettesíti a prompt megfelelő részébe.
- A modell ezután már nem “üresen”, hanem a visszakeresett kontextus alapján válaszol.

Ebben a példában a dokumentumdarabokat egy közös promptba helyezzük be, ezért a modell egyszerre látja a kiválasztott forrásokat és a felhasználó kérdését.

**Kérdések feltevése:**

```python
# Kérdések listája
kerdesek = [
    "Ki volt Nemecsek?",
    "Mi a gittegylet szerepe a történetben?",
    "Ki volt a vörösingesek vezetője?",
    "Mi történt Nemecsekkel a történet végén?"
]

for i, kerdes in enumerate(kerdesek[:2], 1):
    print(f'\n{"="*60}')
    print(f'KÉRDÉS {i}: {kerdes}')
    print('='*60)

    eredmeny = qa_chain.invoke({"input": kerdes})

    print('\nVÁLASZ:')
    print(eredmeny["answer"])

    print('\nFORRÁS DOKUMENTUMOK (első 2):')
    for j, doc in enumerate(eredmeny["context"][:2], 1):
        print(f'\n{j}. {doc.page_content[:150]}...')

    print('\n' + '-'*60)
```

**Hogyan működik a RAG lánc?**

1. Érkezik egy kérdés.
2. A retriever megkeresi a leginkább releváns dokumentumdarabokat.
3. Ezek a darabok bekerülnek a prompt kontextusába.
4. A modell a kérdésre a visszakeresett szövegrészek alapján válaszol.
5. A lánc eredményében megkapjuk:
   - a generált választ (`answer`),
   - valamint a felhasznált dokumentumokat (`context`).

Ez azért hasznos, mert a válasz ellenőrizhetőbbé és dokumentumalapúvá válik.

**Miért jobb ez, mint egy sima LLM hívás?**

Ha a modellt kontextus nélkül kérdezzük meg, akkor általános tudásból próbál válaszolni.  
Ez sok esetben hasznos, de dokumentumalapú kérdéseknél könnyen pontatlan vagy túl általános lehet.

A RAG előnye, hogy:
- a válasz konkrét forrásokra támaszkodik,
- jobban követi a betöltött dokumentum tartalmát,
- kevésbé valószínű, hogy a modell “kitalál” hiányzó részleteket.

```python
# Sima LLM hívás (NINCS RAG)
sima_valasz = llm_puf.invoke("Ki volt Nemecsek?")


print('SIMA LLM VÁLASZ (kontextus nélkül):')
print(sima_valasz)

print('\n' + '='*60)
print('\n💡 Látható különbség:')
print('   - Sima LLM: Általános válasz, lehet pontatlan')
print('   - RAG: Konkrét, a szöveg alapján pontos válasz')
```

## 5. Gyakorlati Feladat: Többlépcsős Tartalomgenerálás

**Cél:** Készítsünk egy kétlépcsős runnable láncot, amely:

1. azonosítja egy megadott téma 4 legfontosabb területét,
2. ezekről rövid, de informatív összefoglalót készít.

A feladat jól szemlélteti, hogyan lehet egy összetettebb problémát több egymásra épülő modellhívásra bontani.

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableLambda

llm_tartalom = ChatOpenAI(
    base_url="https://api.groq.com/openai/v1",
    api_key=api_key,
    model="openai/gpt-oss-120b",
    temperature=0.7,
    max_tokens=1500
)

template_teruletek = """Add meg a '{tema}' témakör 4 legfontosabb területét.
Csak a területek nevét sorold fel, számozva, minden egyéb magyarázat nélkül.

Területek:"""

prompt_teruletek = PromptTemplate.from_template(template_teruletek)

template_kifejtes = """Te egy szakértő vagy. Az alábbi listában megadott témákról írj egy részletes,
de tömör összefoglalót. Minden témát külön bekezdésben fejts ki.

Témák:
{input}

Részletes összefoglaló:"""

prompt_kifejtes = PromptTemplate.from_template(template_kifejtes)

lanc_teruletek = prompt_teruletek | llm_tartalom | StrOutputParser()
lanc_kifejtes = prompt_kifejtes | llm_tartalom | StrOutputParser()

teljes_tartalom_lanc = (
    lanc_teruletek
    | RunnableLambda(lambda text: {"input": text})
    | lanc_kifejtes
)

print("✓ Többlépcsős tartalomgenerálás lánc kész")

```

**Futtatás:**

Most adjunk meg egy témát, és futtassuk le a teljes láncot.

A folyamat:
1. az első lépés előállítja a fő területek listáját,
2. a második lépés ezekből összefoglalót készít,
3. a végeredményt egyetlen hívással kapjuk meg.

Így a lánc több modellműveletet rejt el egy egységes felület mögött.

```python
# Téma megadása
tema = "mesterséges intelligencia"

print(f'Téma: {tema}')
print('\n' + '='*60)

# Lánc futtatása
eredmeny = teljes_tartalom_lanc.invoke({"tema": "mesterséges intelligencia"})

print('\n' + '='*60)
print('VÉGEREDMÉNY:')
print('='*60)
print(eredmeny)
```

**Próbáld ki más témákkal is!**

Érdemes több különböző témát is kipróbálni, például:

- technológiai témák
- társadalmi kérdések
- tudományos területek
- gazdasági vagy környezeti problémák

Így jól megfigyelhető, hogyan változik a generált struktúra és az összefoglalás a bemenet függvényében.

```python
# Más témák kipróbálása
temak = [
    "megújuló energia",
    "kvantumszámítás",
    "gépi tanulás alkalmazásai"
]

for t in temak:
    print(f'\n{"#"*60}')
    print(f'TÉMA: {t.upper()}')
    print('#'*60)
    eredmeny = teljes_tartalom_lanc.invoke(t)
    print(eredmeny)
    print('\n')
```
